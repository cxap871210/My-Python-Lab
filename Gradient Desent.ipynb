```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import random
```

令 Loss funtion $L(x) = x^3 +12x^2 - 5x +3$  微分後 $l(x) = 3x^2+24x -5$

```python
def L(x):
    return x**3 +12*x**2 - 5*x + 3
def l(x):
    return 3*x**2 + 24*x - 5
```

```python
x = np.linspace(-2,2,200)
```

```python
y = x**3 +12*x**2 - 5*x + 3
```

```python
start = random.uniform(-2, 2)
```

```python
start
```

```python
plt.plot(x,y)
plt.scatter(start,L(start),color = 'r')
```

```python
η = 0.005
```

```python
w = start
```

```python
w
```

```python
plt.plot(x,y)
for i in range(50):
    w = w - η * l(w)
    plt.scatter(w,L(w))
```
